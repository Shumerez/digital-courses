# Functions, modules and libraries

1 task

```python
def Discriminant(b, c, a = 1):
    if a!= 0:

        discr = b ** 2 - 4 * a * c
        return discr

    else:
        print('a должно быть отлично от нуля!')

print(Discriminant(3, -2))
```

4 task

```python
def get_sum_odd(a, b):
  sum = 0
  i = a
  while i <= b:
    if i % 2 != 0:
      sum += i
    i += 1
    
  return sum

get_sum_odd(3, 7)
```

# Unit testing

1 task

```python
def grader(score):
  if score < 60:
    return("unsatisactory")
  elif score >= 60 and score <= 74:
    return("satisfactory")
  elif score > 74 and score <= 90:
    return("good")
  else:
    return("excellent")

grader(94)
```

2 task

```python
def solver(a, b):
  if a == 0:
    if b == 0:
      return("any")
    else:
      return("error")
  else:
    return(b / a)

solver(2, 2)
```

# Dictionaries

1 task

```python
n = ['Kate', 'Lena', 'Dima']
a = [23, 56, 33]

def gen_dict(Name, Age):
  d = dict()
  for i in range(len(Name)):
    d[Name[i]] = Age[i]

  return d

gen_dict(n, a)
```
