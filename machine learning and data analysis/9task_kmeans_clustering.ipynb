## Task

A table of some synthetic data is available to you, on the basis of which it is necessary to perform clustering into $K=3$ clusters using the K-means method.  
  
image.png

In sklearn, use initialization with the following centroid coordinates and parameters:  
*KMeans(n_clusters=3, init=np.array([[12.2, 11.4], [15.75, 12.25], [13.67, 9.67]]), max_iter=100, n_init=1)*  
  
Specify to which cluster this or that object will be assigned as a result of clustering (the task is counted only if all clusters are specified correctly).

```python
import pandas as pd
import numpy as np

df = pd.read_csv("drive/MyDrive/data/kmeans_clustering_9task_data.csv", sep=",", decimal='.', index_col='Object')
df.head()
```

```python
km = KMeans(n_clusters=3, init = np.array([[12.2, 11.4], [15.75, 12.25], [13.67, 9.67]]), max_iter=100, n_init=1)
pred = km.fit_predict(df.drop(columns = "Cluster"))
for i in range(len(pred)):
    print("object №%d: cluster %d"%(i + 1, pred[i]))
```

Also, based on the results of clustering, determine the average distances between objects and the centroid related to the cluster $0$

```python
counter = 0
distances = 0
centroid0 = km.cluster_centers_[0]

for i in range(len(pred)):
    if (pred[i] == 0):
        counter += 1
        distances += np.linalg.norm(centroid0 - np.array([df.X.loc[i+1], df.Y.loc[i+1]]))
print(round(distances / counter, 3))
```
