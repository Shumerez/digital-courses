```python
import numpy as np
import pandas as pd
```

```python
m_r = pd.read_csv('m_r.csv')
p_r = pd.read_csv('p_r.csv')
```

```python
m_cd = np.cumsum(m_r['DISTANCE_BACK'])
m_r['CUM_SUM_DIST'] = m_cd
m_r = m_r.sort_values(by=['ID_STOP'])

p_cd = np.cumsum(p_r['DISTANCE_BACK'])
p_r['CUM_SUM_DIST'] = p_cd
p_r = p_r.sort_values(by=['ID_STOP'])
```
